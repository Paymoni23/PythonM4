## Part 2

Please recreate the bar chart seen below.  

* The figure should be 8x6
* The two colors used are `#74b9ff` and `#7ed6df`



```python
# Use this data to generate the plot!
resorts = ["Crested Butte", "Aspen", "Telluride", "Wolf Creek"]
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]
```

```python
plt.figure(figsize=(8,6))
plt.bar(resorts, snow_last_24, color="#74b9ff", label="Last 24hrs")
plt.bar(resorts,snow_next_24 , color="#7ed6df",label="Next 24hrs",bottom= snow_last_24)
plt.ylabel("Inches")
plt.legend()
plt.title("CO Snow Storm Report")
plt.show()
```
