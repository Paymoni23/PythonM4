## Drawing Multiple Plots in One Figure/ Multiple Figures

```python
nums = [1,2,3,4,5]
nums_squared = [1,4,9,16,25]
nums_cubed = [1,8,27,64,125]
```

```python
plt.plot(nums,nums)
plt.plot(nums,nums_squared)
plt.plot(nums,nums_cubed)
```

```python
plt.figure()
plt.plot(nums,nums)
plt.figure()
plt.plot(nums, nums_squared)
plt.plot(nums, nums_cubed)
```

## Bar Plots

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]  
germinated = [74, 88, 56,69,59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, germinated, color="orchid")
plt.title("Germination Rates")
plt.ylabel("Number Of Seeldings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="red")
plt.title("Failure Rates")
plt.ylabel("Number Of Dead Seedling")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db")
plt.bar(plants, germinated, width=0.4, color="#2ecc71", align="edge")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db", bottom=germinated )
plt.bar(plants, germinated, color="#2ecc71")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="#3498db", bottom=germinated,label="Failed Seedlings")
plt.bar(plants, germinated, color="#2ecc71", label="Germinated Seedlings")
plt.legend(facecolor="black")
plt.title("Greenhouse Seedlings")
plt.show()
```

## Subplots

This section is about creating multiple plots within a single figure.

```python
nums = [1,2,3,4,5]
nums_squared = [1,4,9,16,25]
nums_cubed = [1,8,27,64,125]
plt.figure(figsize=(10,4))
plt.suptitle("Our First Subplot", fontsize=30)

plt.subplot(1,3,1)
plt.title("X")
plt.plot(nums, nums)

plt.subplot(1,3,2)
plt.plot(nums, nums_squared, color="red")
plt.title("X Squared")


plt.subplot(1,3,3)
plt.plot(nums, nums_cubed, color="yellow")
plt.title("X Cubed")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4,10))
plt.suptitle("Our Second Subplot", fontsize=30)

plt.subplot(3,1,1)
plt.title("X")
plt.plot(nums, nums)

plt.subplot(3,1,2)
plt.plot(nums, nums_squared, color="red")
plt.title("X Squared")


plt.subplot(3,1,3)
plt.plot(nums, nums_cubed, color="yellow")
plt.title("X Cubed")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.plot(nums)

plt.subplot(2,2,2)
plt.plot(nums)


plt.subplot(2,2,3)
plt.plot(nums)


plt.subplot(2,2,4)
plt.plot(nums)
```
